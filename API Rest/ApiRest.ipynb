```python
#Qual é o padrão arquitetural do REST API?

#O padrão arquitetural do Rest API consiste em uma aplicação simples e padronizada de comunicação
#permitindo que o serviço seja escalonmado e tenha um alto desempenho.
```

```python
#Por que é chamado de "Sossegado" (RESTful)?

#É chamado de "Sossegado" devido de tratar de uma aplicação de comunicação onde há sempre uma conversa entre
#cliente e servidor, onde o servidor recebe uma solicitação de um cliente e devolve uma resposta.
```

```python
#O que é “CRUD”?

#CRUD é o termo em inglês que siginifica Criar, Ler, Atualizar e Excluir, comandos da criação de registros
#para a utilização de uma base de dados em um servidor. O método CRUD é responsavel por facilitar a resposta
#trazida pela REST API
```

```python
#Aprendemos o método GET, qual a correspondência de métodos com o CRUD?

#O método GET possui como correspondência o Read (Ler) no método CRUD, responsavel por ler um registro ou um conjunto
#de registros e devolver uma informação.
```

```python
#Além do método quais os outros três itens de um Request?

#Os outros métodos de requisição são o Post, Put e o Delete.
```
